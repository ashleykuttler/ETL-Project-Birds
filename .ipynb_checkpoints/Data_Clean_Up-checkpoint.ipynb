# Create Species Key

```python
# Read csv file into pandas dataframe
csv_file = "Data/Pop.EstByState.csv"
df = pd.read_csv(csv_file)
df = pd.DataFrame(df)
```

```python
#Select species columns and drop duplicates
df = df[['Sequence AOS 59', 'English Name', 'Scientific Name']]
df=df.drop_duplicates()
df = df.rename(columns={'Sequence AOS 59': 'SPECIES_ID' })
# Create key index
species_key = df.set_index(['SPECIES_ID'])
species_key.head()
```

# CLEAN UP BIRD STRIKE DATA

```python
# Read csv file into pandas dataframe
csv_file = "Data/IL_airplane_strikes_data_2000-2016.csv"
df = pd.read_csv(csv_file, encoding = 'ISO-8859–1')
df = pd.DataFrame(df)
```

```python
#filter for chicago airports and years 2004-2016
df = df[(df['AIRPORT_ID'] == 'KORD') | (df['AIRPORT_ID'] == 'KMDW')]
df = df[(df['INCIDENT_YEAR'] >= 2004) & (df['INCIDENT_YEAR'] <= 2016)]

#keep neccesary columns
bird_strikes = df[['INCIDENT_DATE', 'INCIDENT_MONTH', 'INCIDENT_YEAR', 
         'TIME_OF_DAY', 'TIME','AIRPORT', 'SPECIES', 'COST_REPAIRS', 
         'EFFECT', 'SKY', 'PRECIP', 'BIRDS_SEEN', 'BIRDS_STRUCK', 'SIZE']]

bird_strikes.head()
```
